```python
# clone_voice.py

from bark import generate_audio, preload_models
from scipy.io.wavfile import write
import numpy as np

# Step 1: Preload Bark models
preload_models()

# Step 2: Your reference voice
reference_audio = "ravi_voice.mp3"  # 2-minute audio of your voice

# Step 3: Text you want to speak in your voice
text_to_speak = """
Hello, I am Ravikumar D.
I am an AI and Machine Learning enthusiast.
This is my personal assistant speaking in my voice.
It can answer questions and provide information about me.
"""

# Step 4: Generate audio
audio_array = generate_audio(text_to_speak, history_prompt=reference_audio)

# Step 5: Save audio to WAV file
output_file = "output_cloned_voice.wav"
write(output_file, 22050, np.array(audio_array))

print(f"âœ… Voice cloned successfully! Saved as {output_file}")
```
